```python
# 여기에서부터 입력하세요.


import os
import glob
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

import tensorflow as tf
from tensorflow import keras
from keras.preprocessing.image import ImageDataGenerator



```

```python
video1 = cv2.VideoCapture(TRAIN_PATH + '/220206.mp4')
v1_width = video1.get(cv2.CAP_PROP_FRAME_WIDTH)
v1_height = video1.get(cv2.CAP_PROP_FRAME_HEIGHT)
v1_length = video1.get(cv2.CAP_PROP_FRAME_COUNT)
v1_fps = video1.get(cv2.CAP_PROP_FPS)

print("가로 : ", v1_width)
print("세로 : ", v1_height)
print("길이 : ", v1_length)
print("FPS : ", v1_fps)
video1.release()
```

```python
VIDEO_IMAGE1 = TRAIN_PATH + "/image1"

TIME_MEASUERMENT_UNIT = 10 #TIME MEASUREMENT UNIT을 통해 몇 초 단위로 이미지를 저장할 지 선택

if not os.path.exists(VIDEO_IMAGE1):
  os.mkdir(VIDEO_IMAGE1)

video1 = cv2.VideoCapture(TRAIN_VIDEO1)

while video1.isOpened():
  ret,frame = video1.read()
  if ret:
    # 현재 프레임 위치 (msec) 
    frame_sec = video1.get(cv2.CAP_PROP_POS_MSEC)/1000
    if frame_sec.is_integer():
      if (frame_sec % TIME_MEASUERMENT_UNIT == 0):
        filename = TRAIN_PATH + "/image1/" + str(round(frame_sec)) + ".jpg"
        cv2.imwrite(filename, frame) 
  else:
    break

video1.release()
```

```python
VIDEO_IMAGE2 = TRAIN_PATH + "/image2"

TIME_MEASUERMENT_UNIT = 10 #TIME MEASUREMENT UNIT을 통해 몇 초 단위로 이미지를 저장할 지 선택

if not os.path.exists(VIDEO_IMAGE2):
  os.mkdir(VIDEO_IMAGE2)

video2 = cv2.VideoCapture(TRAIN_VIDEO2)

while video2.isOpened():
  ret,frame = video2.read()
  if ret:
    # 현재 프레임 위치 (msec) 
    frame_sec = video2.get(cv2.CAP_PROP_POS_MSEC)/1000
    if frame_sec.is_integer():
      if (frame_sec % TIME_MEASUERMENT_UNIT == 0):
        filename = VIDEO_IMAGE2 + "/" + str(round(frame_sec)) + ".jpg"
        cv2.imwrite(filename, frame) 
  else:
    break

video2.release()
```

```python
def my_preprocessing(TEST_VIDEO):

  batch_size = 16
  img_height = 96
  img_width = 320

  # 1. VIDEO를 10초 단위로 프레임 추출.
  TIME_MEASUERMENT_UNIT = 10
  video = cv2.VideoCapture(TRAIN_VIDEO)
  while video.isOpened():
    ret, frame = video.read()
    if ret:
      frame_sec = video.get(cv2.CAP_PROP_POS_MSEC)/1000
      if frame_sec.is_integer():
        if (frame_sec % TIME_MEASUERMENT_UNIT == 0):
          filename = TRAIN_IMAGE + '/' + str(round(frame_sec)) + '.jpg'
          cv2.imwrite(filename, frame)
    else:
      break
  video.release()

  # 2. test_generator 생성
  test_datagen = ImageDataGenerator(
      rescale=1. / 255,
  )
  test_generator = test_datagen.flow_from_directory(
      TEST_PATH,
      batch_size = batch_size,
      target_size = (img_height, img_width),
      class_mode = 'categorical',
  )

  return test_generator
```
